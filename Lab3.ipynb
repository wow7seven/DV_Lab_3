# Question 2

```python
flights = pd.read_csv('/content/Flightschedule.csv')
flights
```

### Preprocess

```python
count = flights['Origin'].value_counts().to_dict()
count2 = flights['Destination'].value_counts().to_dict()

for key in count2:
    if key in count:
        count[key] += count2[key]
    else:
        count[key] = count2[key]
```

```python
origins = np.array(flights['Origin'])
origins = np.unique(origins)
origins
```

```python
destinations = np.array(flights['Destination'])
destinations = np.unique(destinations)
destinations
```

```python
cities = list(set(origins) | set(destinations))
cities.sort()
cities
```

```python
foriegn = ['Abu Dhabi', 'Bangkok', 'Dammam', 'Dubai', 'Kuwait', 'Male', 'Muscat', 'Phuket']
```

```python
flights = flights[~flights['Origin'].isin(foriegn) & ~flights['Destination'].isin(foriegn)].reset_index(drop=True)
flights['Origin'] = flights['Origin'].replace('Goa-New Goa', 'Goa')
flights['Destination'] = flights['Destination'].replace('Goa-New Goa', 'Goa')
flights
```

```python
indian_origins = np.array(flights['Origin'])
indian_origins = np.unique(indian_origins)

indian_destinations = np.array(flights['Destination'])
indian_destinations = np.unique(indian_destinations)

indian_cities = list(set(indian_origins) | set(indian_destinations))
indian_cities.sort()
indian_cities
```

```python
flights = flights.drop_duplicates()
flights
```

```python
import geopandas as gpd
from geopandas.tools import geocode


# Geocode the cities to retrieve their latitude and longitude
geocoded_cities = geocode(indian_cities, provider='nominatim')

# Create a dictionary of cities and their latitude and longitude
city_lat_long = {}
for index, row in geocoded_cities.iterrows():
    city_name = row['address'].split(',')[0]
    city_lat_long[city_name] = (row['geometry'].y, row['geometry'].x)

# Print the dictionary of cities and their latitude and longitude
print(city_lat_long)
```

```python
l = list(city_lat_lon['Thoise'])
l
```

### Graph

```python
import csv
import folium

india_map = folium.Map(location=[20, 77], zoom_start=4.2)


for i, row in flights.iterrows():
    folium.PolyLine(
        locations=[ list(city_lat_lon[row['Origin']]) , list(city_lat_lon[row['Destination']])],
        color="blue",
        weight=.25
    ).add_to(india_map)

    folium.CircleMarker(location=list(city_lat_lon[row['Origin']]),
    fill='false',
    radius = count[row['Origin']]/40,
    fill_color='red',
    color = 'clear',
    tooltip = row['Origin'],
    fill_opacity=0.1).add_to(india_map)


india_map
```

```python
import csv
import folium

india_map = folium.Map(location=[20, 77], zoom_start=4.2, tiles="CartoDB dark_matter")


for i, row in flights.iterrows():
    folium.PolyLine(
        locations=[ list(city_lat_lon[row['Origin']]) , list(city_lat_lon[row['Destination']])],
        color="blue",
        weight=.2
    ).add_to(india_map)

    folium.CircleMarker(location=list(city_lat_lon[row['Origin']]),
    fill='false',
    radius = count[row['Origin']]/40,
    fill_color='yellow',
    color = 'clear',
    tooltip = row['Origin'],
    fill_opacity=0.1).add_to(india_map)


india_map
```
